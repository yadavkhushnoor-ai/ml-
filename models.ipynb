```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

X = np.array([
    [2, 2],
    [3, 3],
    [2, 3],
    [8, 7],
    [9, 8],
    [8, 9]
])

y = [0, 0, 0, 1, 1, 1]

model = svm.SVC(kernel='linear')
model.fit(X, y)

w = model.coef_[0]
b = model.intercept_[0]

print("Weights (w):", w)
print("Bias (b):", b)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
ax = plt.gca()
xlim = ax.get_xlim()

xx = np.linspace(xlim[0], xlim[1])
yy = -(w[0] * xx + b) / w[1]

plt.plot(xx, yy, 'k-', label='Hyperplane')
margin = 1 / np.sqrt(np.sum(w**2))
yy_down = yy - np.sqrt(1 + (w[0]/w[1])**2) * margin
yy_up = yy + np.sqrt(1 + (w[0]/w[1])**2) * margin

plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
plt.scatter(model.support_vectors_[:, 0],
            model.support_vectors_[:, 1],
            s=100, facecolors='none', edgecolors='k')

plt.legend()
plt.title("SVM Hyperplane in Colab")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

X = np.array([
    [1, 2],
    [2, 3],
    [2, 1],
    [6, 5],
    [7, 7],
    [8, 6]
])

y = [0, 0, 0, 1, 1, 1]
model = SVC(kernel='linear')
model.fit(X, y)
w = model.coef_[0]
b = model.intercept_[0]

print("w =", w)
print("b =", b)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
x_vals = np.linspace(0, 10, 100)
y_vals = -(w[0]*x_vals + b) / w[1]
plt.plot(x_vals, y_vals, 'k-', label='Hyperplane')
margin = 1 / np.linalg.norm(w)
y_margin_up = y_vals + margin
y_margin_down = y_vals - margin

plt.plot(x_vals, y_margin_up, 'k--')
plt.plot(x_vals, y_margin_down, 'k--')
plt.scatter(model.support_vectors_[:, 0],
            model.support_vectors_[:, 1],
            s=120, edgecolors='k', facecolors='none')

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("2D SVM Hyperplane")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from mpl_toolkits.mplot3d import Axes3D
X = np.array([
    [1, 2, 1],
    [2, 3, 2],
    [2, 1, 2],
    [6, 5, 6],
    [7, 7, 7],
    [8, 6, 7]
])

y = [0, 0, 0, 1, 1, 1]
model = SVC(kernel='linear')
model.fit(X, y)
w = model.coef_[0]
b = model.intercept_[0]

print("w =", w)
print("b =", b)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='bwr')
xx, yy = np.meshgrid(np.linspace(0,10,10), np.linspace(0,10,10))
zz = -(w[0]*xx + w[1]*yy + b) / w[2]
ax.plot_surface(xx, yy, zz, alpha=0.3)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

plt.title("3D SVM Hyperplane (Plane)")
plt.show()
```
